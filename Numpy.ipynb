##Numpy
NumPy (Numerical Python) is a foundational library for scientific computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a vast collection of mathematical functions to operate on these arrays efficiently.It is widely used in fields such as data science, machine learning, engineering, and research due to its ability to handle large datasets and complex mathematical computations with ease. In addition to array operations, NumPy offers tools for linear algebra, random number generation, Fourier transforms, and more, making it an essential tool for anyone working with numerical data in Python.

```python
#import the library
import numpy as np
```

```python
# ndarray
x = np.array([1,2,3,4]) #1 dimentional
print(x)

y = np.array([[1.0,'h',3],[4,5,6]])
print(y)
print(type(y))
print(y.dtype)
print(y.shape)
# u32 unicode string
```

```python
# how to make a zero matrix, ones matrix, diagonal matrix
arr1 = np.zeros((3,4))
# print(arr1)

arr2 = np.ones((2,2))
# print(arr2)

arr3 = np.eye(6)
print(arr3)

arr4 = np.diag([10, 20, 30])
print(arr4)

# full()
```

```python
# arange(start, stop, step)
z = np.arange(10)
print(z)
print(type(z))

#reshape
z = z.reshape((2,5))
print(z)

# random
a = np.random.randint(0,5, (2,3))
print(a)
# random.random
b = np.random.random((2,3)) #range [0,1)
print(b)

# slicing
print(z[0,0:5:2])
print(z[1])
print(z[1,2])
```

# NumPy Mathematical Functions

| Function | Description | Usage Example |
|----------|-------------|---------------|
| **Addition** | Element-wise addition of two arrays | `np.add(X, Y)` or `X + Y` |
| **Subtraction** | Element-wise subtraction of two arrays | `np.subtract(X, Y)` or `X - Y` |
| **Multiplication** | Element-wise multiplication of two arrays | `np.multiply(X, Y)` or `X * Y` |
| **Division** | Element-wise division of two arrays | `np.divide(X, Y)` or `X / Y` |
| **Exponential** | Computes e^x for each element | `np.exp(x)` |
| **Square Root** | Computes square root of each element | `np.sqrt(x)` |
| **Power** | Raises each element to specified power | `np.power(x, 2)` |
| **Mean** | Calculates average of array elements | `np.mean(array)` |
| **Sum** | Calculates sum of all array elements | `np.sum(array)` |
| **Median** | Finds median (middle value) of array | `np.median(array)` |
| **Maximum** | Finds maximum value in array | `np.max(array)` |
| **Minimum** | Finds minimum value in array | `np.min(array)` |
| **Standard Deviation** | Calculates standard deviation of array | `np.std(array)` |

```python
x = np.array([[1,2,3,4],
             [2,3,2,3]]) #2x4

y = np.array([[5,2,1,1],
              [6, 3,2,2],
              ])#4x2
print(np.divide(x,y))
# print(np.dot(x,y))
print(np.max(x))
```

##Pandas
Based on the term Panel Data from econometrics, the Python library Pandas was developed for efficient data manipulation and analysis. It introduces two core data structures: the Pandas Series and the Pandas DataFrame. These structures are specifically designed to be flexible, easy to use, and optimized for fast data operations. A common challenge in real-world data is the presence of issues like missing values, often represented as Not a Number (NaN), as well as outliers and incorrect entries. Effectively cleaning and preparing such data is essential, as a dataset with numerous missing or erroneous values will prevent machine learning algorithms from performing well.

##Pandas series
A Pandas Series is a one-dimensional, labeled array that can hold data of any type (integers, strings, floating-point numbers, Python objects, etc.). It is one of the core data structures in the Pandas library. The labels, collectively known as the index, allow for fast and intuitive data access, alignment, and manipulation. Essentially, a Series combines the functionality of a list and a dictionary, providing powerful, vectorized operations for efficient data analysis.

```python
import pandas as pd
# how to define a pandas series
fruits = pd.Series(data = [1,20,15,2],index = ['orange','banana', 'apple','kiwi'])
pd1 = pd.Series(data = [10,23,12,20],index = ['o','banana', 'apple','kiwi'])
fruits
print(fruits[['banana', 'apple']])
fruits[['banana', 'apple']]
print(fruits.iloc[0:2])
print(fruits.loc['orange':'apple'])
print(fruits.ndim)
print(fruits.values)
print(fruits.index)
print(fruits.size)
print(fruits.shape)

# arithmetic operations (+,  -,*, /) on Pandas Series.
print(fruits+pd1)
# null NaN None

# with numpy
print(np.add(fruits,pd1))
```

##DataFrame
A DataFrame is a two-dimensional, labeled data structure in Pandas, resembling a spreadsheet or a SQL table. It consists of rows and columns, where each column is a Pandas Series sharing a common index. DataFrames are designed to store and manipulate tabular data, offering a powerful and flexible toolkit for tasks like data cleaning, transformation, merging, filtering, and aggregation, making them the fundamental workhorse for data analysis in Python.


```python
# Create a DataFrame Manually
# First way: create a dictionary of Pandas Series.
fruits = {
    'store1': pd.Series(data = [1,20,15,2],index = ['orange','banana', 'apple','kiwi']),
    'store2': pd.Series(data = [10,23,12,20],index = ['o','b', 'apple','kiwi'])
}
df = pd.DataFrame(fruits)
print(df)
# Second way: create a dictionary of lists.
fruits = {
    'store1': ['orange','banana', 'apple','kiwi'],
    'store2': ['orange','banana', 'apple','kiwi']
}
df = pd.DataFrame(fruits)
df
# Third way: create a list of dictionaries.
fruits = [
    {1:'a',
     2: 'b',
     3: 'c',
     4: 'd'},
    {3:'a',
     4: 'b'}
]
df = pd.DataFrame(fruits)
print(df)

#  access elements
df[4]
print(df.ndim)
print(df.values)
print(df.index)
print(df.size)
print(df.shape)
# add col to exiting dataframe
df[5] = ['s','c']
df
# Add a new column based on the arithmetic operation between existing columns of a DataFrame
df['store3'] = np.add(df['store1'],df['store2'])
df



# Delete multiple columns from a DataFrame using .drop()
df = df.drop(['store3'], axis=1)
df
# Delete rows from a DataFrame using .drop()
df = df.drop(axis = 0)
df

# groupby()
d = df.groupby('store1')['store2'].mean()
print(d)
df
# Count the total NaN values
count_nan = df.isnull().sum().sum()
print(count_nan)
# Drop rows having NaN values
df = df.dropna(axis = 0)
df
# Drop columns having NaN values
df = df.dropna(axis = 1)
df
# Replace NaN
df['store1'] = df['store1'].fillna(np.mean(df['store1']))
df

```
