##Escape Sequence
In Python, an escape sequence is a series of characters beginning with a backslash `\` that represents a special character or action within a string. They are used to include characters that are difficult or impossible to type directly, such as newlines, tabs, or quotes that would otherwise terminate the string.
| Escape Sequence | Meaning |
|-----------------|---------|
| `\\`            | Prints a backslash (`\`) |
| `\'`            | Prints a single quote (`'`) |
| `\"`            | Prints a double quote (`"`) |
| `\n`            | Moves to a new line |
| `\t`            | Adds a tab space (like pressing the Tab key) |



```python
print('he\"s smart')
print('new\nline')
print('12\t34')
print('yes\\no')
```

##Variables and Data Types
Variables are used to store data for future use. Think of a variable as a box with a name where you can keep information. Whenever you want to use that information in your code, you simply call the box by its name.

When naming a variable, Python follows specific rules:
- ✅ Must start with a letter (A–Z, a–z) or an underscore _  
  Examples: first_name, _counter
- ❌ Cannot start with a number                                                  
  Example: 2name is invalid
- ✅ Can include letters, numbers, and underscores                              
  Examples: user1, total_amount
- ❌ Cannot use Python reserved keywords           
  These are part of the language syntax and can't be used as variable names

  In Python, every variable has a data type that reflects the kind of value stored inside it. For example, if a variable contains text its type will be a string, if it contains a number its type will be an integer or float, . Unlike some programming languages, Python does not require you to explicitly declare the data type when creating a variable. Instead, Python automatically infers the type based on the value you assign, which makes the language flexible and easy to use.



| Data Type | Description |
|-----------|-------------|
| **int**   | Represents whole numbers without decimals, e.g., `5`, `-12`, `1000`. |
| **float** | Represents numbers with decimals (floating-point), e.g., `3.14`, `-0.5`, `2.0`. |
| **bool**  | Represents logical values: `True` or `False`. |
| **string**| Represents sequences of characters (text), e.g., `"Hello"`, `'Python'`. |


```python
num= 12
num1 = 12.2
t = True #1 2 3
f = False # 0 0.0
name = 'na'
```

##Type Checking and Conversion

In Python, the `type()` function is used to check the data type of any variable. It helps you understand what kind of value a variable holds, whether it's an integer, float, string, boolean, or another type. For example, type(5) returns <class 'int'>, and type("hello") returns <class 'str'>. If you want to convert between data types, Python provides built-in functions like `int()`, `float()`, `str()`, and `bool()`. These allow you to change a value from one type to another.




```python
name = 'naheel'
print(type(name))
num = 12.
num = int(num)
b = -1
b = bool(b)
print(b)
print(type(b))
```

##String Methods
There are a lot of methods that belong to the string data type that help us when we deal with strings. For example, the `lower()` method changes all the letters in a string to lowercase, and the `upper()` method changes them to uppercase. The `capitalize()` method changes the first letter of a string to uppercase and the rest to lowercase. These methods are useful when you want to standardize or format text. Together, these string methods make it easy to work with words and sentences in your programs.

```python
s = 'Hello World'
print(s.lower())
print(s.upper())
print(s.capitalize())
```

##Arithmetic Operators

| Operator | Description |

##Assignment Operators
Assignment operators are used to give values to variables and update them in different ways. The simplest one is =, which assigns a value directly. Others like +=, -=, *=, and /= let you update the value of a variable by combining assignment with an operation.

| Operator | Name                     | Example                     |
|----------|--------------------------|-----------------------------|
| `=`      | Assignment Operator      | `a = 7`                     |
| `+=`     | Addition Assignment      | `a += 1   # a = a + 1`      |
| `-=`     | Subtraction Assignment   | `a -= 3   # a = a - 3`      |
| `*=`     | Multiplication Assignment| `a *= 4   # a = a * 4`      |
| `/=`     | Division Assignment      | `a /= 3   # a = a / 3`      |
| `%=`     | Remainder Assignment     | `a %= 10  # a = a % 10`     |
| `**=`    | Exponent Assignment      | `a **= 10 # a = a ** 10`    |


```python
n = 10
n//=2
print(n)
```
