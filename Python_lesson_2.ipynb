##Conditional Statements

Conditional statements are used to control the flow of a program by allowing decisions to be made based on specific conditions. The most common keywords are `if`, `elif`, and `else`, which let the program execute different blocks of code depending on whether a condition evaluates to true or false. For example, an if statement checks a condition and runs its block if the condition is true, while elif provides additional checks, and else handles all remaining cases. These statements are essential for building logic in programs, enabling developers to create dynamic and responsive code that adapts to different inputs or situations.

# Comparison Operators

| !=       | Not equal                | x != y  |
| >        | Greater than             | x > y   |

# Logical Operators

Logical operators are used to combine conditional statements.

| Operator | Description                                      | Example                  |
|----------|--------------------------------------------------|--------------------------|
| and      | Returns True if both statements are true         | x < 5 and x < 10         |
| or       | Returns True if one of the statements is true    | x < 5 or x < 4           |
| not      | Reverses the result; returns False if true       | not(x < 5 and x < 10)    |


```python
t = -15
if t>=0 or t<=35:
  print('the wether is good')
else:
  print('the wether is not good')
```

##Nested if
We use nested if statements when we want to check several conditions and handle multiple branches of logic. In other words, it allows us to place one if statement inside another, so that a condition is only checked if the previous one is true.

```python
a =2
b= 3
c = -4
max = None
if a>=b:
  if a>=c:
    max = a
  else:
    max = c
else:
  if b>=c:
    max=b
  else:
    max= c
print(max)
```

##While Loop
A while loop is used to repeatedly execute a specific block of code as long as a given condition remains true. In other words, the loop keeps running until a certain event occurs or the condition becomes false.

```python
x =1
while x<=5:
  print('hello world')
  print(x)
  x+=1
```

```python
x = 5
flag = True
while flag:
  print(x)
  x-=2
  if x<=0:
    flag = False
```

##break and continue
Sometimes, when we are running a loop, we want to control how it behaves by using:                                                                               
break - Used to stop the loop completely when a certain condition is met.                                                                         
continue - Used to skip the rest of the code in the current loop iteration and go back to the beginning of the loop for the next iteration.






```python
x = 0
while x<10:
  x+=1
  if x == 5:
    continue
  print(x)
```

##For Loop
A for loop is used when we want to repeat a specific block of code a certain number of times. In other words, we use it when we already know how many times the code should run. This makes the for loop especially useful for working with sequences such as lists, strings, or ranges of numbers.

```python
text = 'abc'
for i in text:
  print(i)
```

```python
name = 'hello world'
for letter in name:
  print(letter, end = ' ')
```

##Range
The built-in range() function returns an immutable sequence of numbers, commonly used for looping a specific number of times.

This set of numbers has its own data type called range.


```python
# range(start, stop, step)
x = range(1,10, 2) #0 1 2 3
print(type(x))
range(3) # start = 0 , end = 3
range(2, 6)# start = 2 , end = 6
range(1 , 10 , 3)
print(len(x))
for num in range(1,10, 2):
  print(num)
```
