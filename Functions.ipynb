##Functions
Functions are reusable blocks of code designed to perform specific tasks. We use them when we want to execute the same block of code multiple times without rewriting it. A function only runs when it is called, and it can optionally return data as a result. By organizing code into functions, we reduce repetition, improve readability, and make programs easier to maintain and debug.
 Python provides many built-in functions ( like `print()`, `len()` and others ) that are ready to use for common tasks. In addition, you can create your own user-defined


```python
# how to define a function
def greet():
  name = input('Enter your name:')
  print('hello ',name)

# Call the function to execute it
greet()
```

```python
# Function with parameters (arguments)
# Arguments: the actual values that are passed to a function when it is called.
# Parameters: the variables listed in the function definition that receive the arguments.

def add(a, b): #argument
  print(a+b)

number_1 = int(input('Enter a number: '))
number_2 = int(input('Enter a number: '))
add(number_1, number_2)
```

```python
# Function that returns a value
def add(a,b):
  result = a+b
  return result


print(add(4,4))
```

```python
# Function that returns multiple values
def add(a,b):
  x = 2
  return a, b, x

num1, num2, num3 = add(1,2)
print(num1)
print(num2)
print(num3)
```

```python
# Function with default parameters
# This function has default values for its parameters
# If no arguments are provided, it uses the defaults

def add(a = 2, b = 4): #defult parameters
  result = a+b
  return result

add(5)

# range is one of the pre_defined functions that uses defult parameters
# range(start = 0, end, step =1)
for i in range(8): # start = 0, end = 8, step = 1
  print(i)

print('\n')

for i in range(5, 8): # start = 5, end = 8, step = 1
  print(i)

print('\n')

for i in range(1, 8,2): # start = 5, end = 8, step = 1
  print(i)
```

```python
# Functions with variable-length arguments (*args)
# The *num parameter allows the function to accept any number of arguments when we call it
def add(*num): #defult parameters
  return sum(num)


# Function that calls another function
def avg(*num):
  return add(*num)/len(num) # Calls add() function

print(add(2,2,3,4))
print(avg(2,3,4,5,6))
```

```python
# Recursive function
# A recursive function calls itself
def a(n):
  if n>=0:
    a(n-1)
    print(n)


a(5)
```

##Documentation
In Python, documentation is used to explain how code works and make it easier to understand, maintain, and reuse. A common way to document code is through docstrings, which are special strings written inside functions using triple quotes (""" ... """). They describe the purpose of the code and can be accessed with tools like help() or the .__doc__ attribute, making them very useful for both developers and learners.

```python
def add(a = 2, b = 4):
  """ this function calculate the sum of two numbers
  input: a: first number can be float or int
         b: second number can be float or int
  output: return the sum of two numbers
  """
  #docstring
  result = a+b
  return result
```

The docstring can be accessed with help(function_name) or function_name.__doc__ or by writing ? after the function name

```python
add?
```

```python
help(add)
```

```python
print(add.__doc__)
```

```python
#we can use the same way to see the docstring for any
#function, even for built-in functions like print()
print?
```

```python
print(print.__doc__)
```

```python
help(print)
```

```python
# Example of using triple quotes for multi-line strings
# Triple quotes can be used for strings that span multiple lines
print('''1
2
3
4
5
''')
```

##Lambda
A lambda function in Python is a small, anonymous function created with the lambda keyword. It is typically used for short, one‑line operations where defining a full function would be unnecessary. Lambdas are often applied in places like sorting, filtering, or quick calculations, helping write compact and efficient code without losing readability.

```python
# how to define a lambda
# Lambda functions are anonymous (no name) and defined in one line

z = lambda a,b: a+b
print(z(3,5))

x = lambda a=1,b=2: a+b # lambda with defult parameters
print(x())

def add(*num):
  return sum(num)

y = lambda *num: add(*num)/len(num)
print(y(1,2,3,4,5))
```
