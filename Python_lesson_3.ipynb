##Data Structures
We have already learned about basic data types such as int, string, and others. Sometimes, however, we need to store multiple types of data together in one place. To achieve this, we use data structures.

Data structures are containers or collections that organize and group different data types in various ways. You can think of them like file folders: each folder holds and organizes different files, making it easier to manage and access information.

##List
A list in Python is a collection that allows us to store multiple items together in a single variable. These items can be of different data types, such as numbers, strings, or even other lists. Lists are ordered, which means the items keep the same sequence in which they were added, and they are also changeable, so we can add, remove, or update elements after creating the list. Unlike some other data structures, lists allow duplicate values, meaning the same item can appear more than once. Each item in a list is accessed by its index, starting from 0 for the first item, 1 for the second, and so on, making it easy to retrieve or modify specific elements.

```python
names = ['naheel', 'nada', 'ali']
numbers = [1, 2, 3]
info = ['nada', 20, True, 3.0]
#index
# 'nada'    0   -4
# 20        1   -3
# True      2  -2
# 3.0       3  -1
print(info[-3])
print(info)
#len()
print(len(info))
print(info[len(info)-1])
# type()
print(type(info))
```

```python
# append()
# info.append('hi')
# # insert()
# info.insert(3, 'a')
# print(info)

# for i in info:
#   print(i)
#remove()
info.remove('a')
print(info)

#extend()
info.extend(numbers)
print(info)
print(numbers)
```

# Membership Operators

Membership operators are used to check whether a value exists in a sequence (like a list or string).

| Operator | Description |
|----------|-------------|
| in       | Returns True if a sequence with the specified value is             present in the object |
| not in   | Returns True if a sequence with the specified value is not present in the object |


```python
for i in list1:
  print(i)

if  10 not in list2:
  print('yes')
else:
  print('no')
```

##Dictionary
This data structure allows us to store data in key-value pairs, where each key is unique and maps to a specific value.

```python
info = {
    'name' : 'naheel',
    'age' : 22,
    'country' : 'Palestine',
    'dict' : {1:1,
              2: 2,
              3: 3},
    'list': [1,2,3]
}

print(info)
print(type(info))
print(info['list'])
print(info.keys())
print(info.values())
info['name'] = 'n'
info['len'] = 160
info[1] = 1
print(info)

for i in info:
  print(i, end = ' ')
```

##Tuple
Tuples are a data structure used to store multiple items in a single variable. They are ordered collections, meaning the items maintain a specific sequence, and they are immutable, which means their contents cannot be changed after creation. Tuples are useful when you want to group related data together and ensure that it remains constant throughout the program.



```python
t = (1,'hi',True)
print(t[2])
#slice
print(t[0:2])
# join
s = ('a', 'b', 'c')
print(' '.join(s))
s = 'a', 'b', 'c'
print(type(s))

a, b, c = 1,2,3
print(c)
```

##Set
Sets are a data structure used to store multiple items in a single variable. Unlike lists or tuples, sets are unordered, meaning the items do not maintain a specific sequence. They are also unindexed, so you cannot access elements by position. Additionally, sets are immutable in terms of structure - you cannot change individual elements once the set is created, although you can add or remove entire items. Sets are commonly used to store unique values and perform operations like union, intersection, and difference.


```python
set1 = {1,2, 'a', "True"}
print(set1)

name = 'mohammedde'
set2 = set()
for char in name:
  set2.add(char)
set2.pop()

print(set2)
```

```python
set1 = set(range(0,10))
set2 = set(range(0,21))
print(set1.union(set2))
print(set1.intersection(set2))
```
